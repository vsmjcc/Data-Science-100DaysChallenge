# **Red Wine Quality Classification using Decision Tree**

 **1. Data Pre-processing :**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Dataset

df = pd.read_csv('winequality-red.csv')
df
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
# Divide the Dataset into Dependent and Independent Variables

X = df.drop('quality', axis=1)
Y = df['quality']
```

```python
# Train-test-split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.20, random_state=42)
```

 **2. Training the Model :**

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
model = classifier.fit(x_train, y_train)
```

 **3. Predict the Result :**

```python
y_pred = classifier.predict(x_test)
```

```python
pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
```

 **4. Model Evaluation :**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
mat = confusion_matrix(y_test, y_pred)
print(mat)
```

 **5. Plot the Decision Tree :**

```python
from sklearn import tree
```

```python
plt.figure(figsize=(15, 15))
tree.plot_tree(model, filled=True)
```

```python
plt.figure(figsize=(20, 10))
plot_tree(classifier, filled=True, feature_names=X.columns, class_names=['3', '4', '5', '6', '7', '8'])
plt.show()
```

## Decision Tree for IRIS Dataset

 **1. Data Pre-processing :**

```python
df = pd.read_csv("IRIS.csv")
```

```python
df.info()
```

```python
# Get unique values in Obejct data type column

df['species'].unique()
```

```python
# Encoding the String values from the 'Species' column

df['species'] = df['species'].map({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3})
```

```python
df.info()
```

```python
# Divide dataset in dependent and independent variables

X = df.drop('species', axis=1)
Y = df['species']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

 **2. Training the Model :**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier()
model = classifier.fit(x_train, y_train)
```

 **3. Predict the Output :**

```python
y_pred = classifier.predict(x_test)
```

 **4. Model Evaluation Techniques :**

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
rp = classification_report(y_test, y_pred)
print(rp)
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
```

```python
mat = confusion_matrix(y_test, y_pred)
print(mat)
```

 **5. Plot the Decision Tree :**

```python
from sklearn import tree
```

```python
plt.figure(figsize=(15, 10))
tree.plot_tree(model, filled=True)
```
