## Handling Missing Values

```python
# NaN : Missing value
weather=pd.read_csv("weather_data1.csv")
```

```python
weather
```

```python
# Identying missing values
weather.isnull().sum()
```

```python
weather.notnull().sum()
```

```python
# Missing value imputation : Replacing missing values with sensible information
weather.fillna(0)
```

```python
weather.fillna({"temperature":0,"windspeed":0,"event":"No Event"})
```

```python
weather.fillna({"temperature":weather.temperature.mean(),"windspeed":weather.windspeed.mean(),"event":"No Event"})
```

```python
weather.fillna(method="ffill")
```

```python
weather.fillna(method="bfill")
```

```python
weather
```

```python
weather1=weather.set_index(weather.day)
```

```python
weather1
```

```python
weather.info()
```

```python
weather['day']=pd.to_datetime(weather.day)
```

```python
weather.info()
```

```python
weather.set_index(weather.day,inplace=True)
```
