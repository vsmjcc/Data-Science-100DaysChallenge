# **Student Performance Analysis using Multiple Linear Regression**

 **Import Important Libraries :**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

 **Import Dataset :**

```python
df = pd.read_csv('Student_Performance.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

 **EDA :**

```python
df['Extracurricular Activities'].unique()
```

```python
# Encoding categorical values

d = {'Yes':0, 'No':1}
df['Extracurricular Activities'] = df['Extracurricular Activities'].map(d)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

 **Splitting the Dataset :**

```python
X = df.drop('Performance Index', axis=1)
Y = df['Performance Index']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
```

 **Training the Model :**

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

 **Evaluating Results :**

```python
y_pred = regressor.predict(x_test)
```

```python
out = pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
out
```

 **Model Evaluation :**

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2 = r2_score(y_test, y_pred)
r2

# r2_score = 0.9889 ~ 1
# So, the model has good fit to the dataset
```

```python
sq_err = mean_squared_error(y_test, y_pred)
sq_err
```

```python
regressor.coef_
```

# Red Wine Quality

```python
data = pd.read_csv('winequality-red.csv')
print(data.head())
```

```python
data.describe()
```

```python
data.isnull().sum()
```

**Prepare Data:**

```python
X = data[['fixed acidity', 'volatile acidity', 'alcohol']].values
y = data['quality'].values
```

**Split Data into Training and Testing Sets:**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

**Build and Train the Multiple Linear Regression Model:**

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

**Make Predictions:**

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 = df.head(10)
df1
```

```python
df1.plot(kind='bar',figsize=(16,10))
plt.show()
```

**Evaluate the Model:**

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
```

**Interpret the Results:**

```python
coefficients = regressor.coef_
intercept = regressor.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```
